```python
import pandas as pd
import numpy as np
from tqdm import tqdm

def find_timestamp_extremum(data, df_lowest_timeframe):
    """
    :params: data(highest timeframe OHLCV data),
             df_lowest_timeframe (lowest timeframe OHLCV data)
    :return: data with three new columns: Low_time (TimeStamp), 
             High_time (TimeStamp), High_first (Boolean)
    """
    # Create a copy to avoid modifying the original DataFrame
    data = data.copy()
    
    # Identify stock tickers (assumes column names like 'High_STOCK')
    stocks = list(set(col.split('_')[1] for col in data.columns if '_' in col))
    
    # Prepare new columns with explicit datetime type
    data["Low_time"] = pd.NaT
    data["High_time"] = pd.NaT
    data["First"] = np.nan

    # Loop to find out which of the Take Profit and Stop Loss appears first
    for i in tqdm(range(len(data) - 1)):
        # Extract values from the lowest timeframe dataframe
        start = data.iloc[i:i + 1].index[0]
        end = data.iloc[i + 1:i + 2].index[0]
        
        # Initialize lists to store high and low times for all stocks
        high_times = []
        low_times = []
        
        # Process each stock
        for stock in stocks:
            try:
                # Try to find high and low columns for each stock
                high_col = [col for col in df_lowest_timeframe.columns if f'High_{stock}' in col][0]
                low_col = [col for col in df_lowest_timeframe.columns if f'Low_{stock}' in col][0]
                
                # Extract rows for this time period
                row_lowest_timeframe = df_lowest_timeframe.loc[start:end].iloc[:-1]
                
                # Find high and low times
                high = row_lowest_timeframe[high_col].idxmax()
                low = row_lowest_timeframe[low_col].idxmin()
                
                high_times.append(high)
                low_times.append(low)
            
            except Exception as e:
                print(f"Error processing stock {stock}: {e}")
                continue
        
        # If we found any high/low times
        if high_times and low_times:
            # Get the earliest high and lowest low
            earliest_high = min(high_times)
            earliest_low = min(low_times)
            
            # Explicitly convert and set times
            data.at[start, "Low_time"] = pd.Timestamp(earliest_low).tz_localize(None)
            data.at[start, "High_time"] = pd.Timestamp(earliest_high).tz_localize(None)
        else:
            # Fallback if no times found
            data.at[start, "Low_time"] = pd.Timestamp(start)
            data.at[start, "High_time"] = pd.Timestamp(start)

    # Find out which appears first
    data.loc[data["High_time"] > data["Low_time"], "First"] = 1
    data.loc[data["High_time"] < data["Low_time"], "First"] = 2
    data.loc[data["High_time"] == data["Low_time"], "First"] = 0

    # Verify the number of rows without both TP and SL on same time
    percentage_garbage_row = len(data.loc[data["First"] == 0].dropna()) / len(data) * 100
    print(f"WARNINGS: Garbage row: {percentage_garbage_row:.2f}%")

    # We delete the last row because we can't find the extremum
    data = data.iloc[:-1]

    # Remove timestamp column if it exists
    if "timestamp" in data.columns:
        del data["timestamp"]

    return data
```

```python
def run_tp_sl(data, leverage=1, tp=0.015, sl=-0.015, cost=0.00, stock="TSLA"):
    """
    :params (mandatory): data (must contain High_time and Low_time columns)
    :params (optional): leverage=1, tp=0.015, sl=-0.015, cost=0.00
    :return: data with three new columns: Low_time (TimeStamp), High_time (TimeStamp), High_first (Boolean)
    """
    
    # Define the stock column names based on the stock ticker passed
    open_col = f"Open_{stock}"
    high_col = f"High_{stock}"
    low_col = f"Low_{stock}"
    
    # If the columns need to be converted to UTC, use tz_convert
# Handle mixed tz-naive and tz-aware data
    if data['Low_time'].dt.tz is None:
        data['Low_time'] = data['Low_time'].dt.tz_localize('UTC')
    if data['High_time'].dt.tz is None:
        data['High_time'] = data['High_time'].dt.tz_localize('UTC')



    
    
    # Initialize parameters
    buy = False
    sell = False
    data["duration"] = 0

    for i in range(len(data)):
        # Extract current row
        row = data.iloc[i]

        ######## OPEN BUY ########
        if not buy and row["Signal"] == 1:
            buy = True
            open_buy_price = row[open_col]
            open_buy_date = row.name

        # Verify conditions for buy trades
        if buy:
            var_buy_high = (row[high_col] - open_buy_price) / open_buy_price
            var_buy_low = (row[low_col] - open_buy_price) / open_buy_price

            # Check for both TP and SL on the same candle
            if var_buy_high > tp and var_buy_low < sl:
                if row["Low_time"] == row["High_time"]:
                    pass
                elif row["First"] == 2:
                    data.loc[row.name, "returns"] = tp * leverage
                    # Ensure open_buy_date is not None before calculating duration
                    if open_buy_date is not None:
                        data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()
                    else:
                        data.loc[row.name, "duration"] = None  # Or set to a default value


                elif row["First"] == 1:
                    data.loc[row.name, "returns"] = sl * leverage
                    # Ensure open_buy_date is not None before calculating duration
                    if open_buy_date is not None:
                        data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()
                    else:
                        data.loc[row.name, "duration"] = None  # Or set to a default value


                buy = False
                open_buy_price = None
                var_buy_high = 0
                var_buy_low = 0
                open_buy_date = None
            elif var_buy_high > tp:
                data.loc[row.name, "returns"] = tp * leverage
                # Ensure open_buy_date is not None before calculating duration
                if open_buy_date is not None:
                    data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()
                else:
                    data.loc[row.name, "duration"] = None  # Or set to a default value


                buy = False
                open_buy_price = None
                var_buy_high = 0
                var_buy_low = 0
                open_buy_date = None
            elif var_buy_low < sl:
                data.loc[row.name, "returns"] = sl * leverage
                # Ensure open_buy_date is not None before calculating duration
                if open_buy_date is not None:
                    data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()
                else:
                    data.loc[row.name, "duration"] = None  # Or set to a default value


                buy = False
                open_buy_price = None
                var_buy_high = 0
                var_buy_low = 0
                open_buy_date = None

        ######## OPEN SELL ########
        if not sell and row["Signal"] == -1:
            sell = True
            open_sell_price = row[open_col]
            open_sell_date = row.name

        # Verify conditions for sell trades
        if sell:
            var_sell_high = -(row[high_col] - open_sell_price) / open_sell_price
            var_sell_low = -(row[low_col] - open_sell_price) / open_sell_price

            if var_sell_low > tp and var_sell_high < sl:
                if row["Low_time"] == row["High_time"]:
                    pass
                elif row["First"] == 1:
                    data.loc[row.name, "returns"] = tp * leverage
                    data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()

                elif row["First"] == 2:
                    data.loc[row.name, "returns"] = sl * leverage
                    data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()

                sell = False
                open_sell_price = None
                var_sell_high = 0
                var_sell_low = 0
                open_sell_date = None
            elif var_sell_low > tp:
                data.loc[row.name, "returns"] = tp * leverage
                # Ensure open_buy_date is not None before calculating duration
                if open_buy_date is not None:
                    data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()
                else:
                    data.loc[row.name, "duration"] = None  # Or set to a default value


                sell = False
                open_sell_price = None
                var_sell_high = 0
                var_sell_low = 0
                open_sell_date = None
            elif var_sell_high < sl:
                data.loc[row.name, "returns"] = sl * leverage
                # Ensure open_buy_date is not None before calculating duration
                if open_buy_date is not None:
                    data.loc[row.name, "duration"] = (row.Low_time - open_buy_date).total_seconds()
                else:
                    data.loc[row.name, "duration"] = None  # Or set to a default value

                sell = False
                open_sell_price = None
                var_sell_high = 0
                var_sell_low = 0
                open_sell_date = None

    # Fill missing returns with 0
    data["returns"] = data["returns"].fillna(value=0)
    return data
```

```python
import yfinance as yf
import pandas as pd

def resample_ohlcv_data(data_ohlcv, high_timeframe='1h', low_timeframe='15min'):
    """
    Resample OHLCV data for multiple stocks with a generic approach
    
    Parameters:
    -----------
    data_ohlcv : pandas.DataFrame
        Multi-level column DataFrame with OHLCV data
    high_timeframe : str, optional (default='1h')
        Timeframe for higher resolution resampling
    low_timeframe : str, optional (default='15min')
        Timeframe for lower resolution resampling
    
    Returns:
    --------
    tuple : (high_timeframe_data, low_timeframe_data)
    """
    # Get all unique stock tickers
    stocks = set([col[1] for col in data_ohlcv.columns])
    
    # Prepare aggregation dictionaries
    agg_methods = {
        'Open': 'first',
        'High': 'max',
        'Low': 'min',
        'Close': 'last',
        'Adj Close': 'last',
        'Volume': 'sum'
    }
    
    # Create resampled dataframes for each timeframe
    resampled_data = {}
    for timeframe in [high_timeframe, low_timeframe]:
        # Prepare aggregation dictionary for all stocks
        agg_dict = {}
        for stock in stocks:
            for col, method in agg_methods.items():
                full_col_name = (col, stock)
                agg_dict[full_col_name] = method
        
        # Resample the data
        resampled = data_ohlcv.resample(timeframe).agg(agg_dict).dropna()
        
        # Flatten column names if needed
        resampled.columns = ['_'.join(col).strip() for col in resampled.columns]
        
        resampled_data[timeframe] = resampled
    
    return resampled_data[high_timeframe], resampled_data[low_timeframe]


high_timeframe_data, low_timeframe_data = resample_ohlcv_data(data_ohlcv)

# Print the first few rows of each
print("High Timeframe Data:")
print(high_timeframe_data.head())
print("\nLow Timeframe Data:")
print(low_timeframe_data.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.interpolate import interp1d

def monte_carlo(data, method="simple", start_year=2012, end_year=2021, num_simulations=100):
    random_returns = []
    data["returns"] = data["returns"].fillna(value=0)  # Fill NaN values with 0
    
    # Convert "returns" to numpy array to avoid pandas Series warning when shuffling
    returns = np.array(data["returns"])
    
    for _ in tqdm(range(num_simulations)):  # Run the specified number of simulations
        np.random.shuffle(returns)  # Shuffle returns
        random_returns.append(returns.copy())  # Make a copy to avoid modifying the original
        
    if method == "simple":
        # Cumulative sum for simple returns
        df_ret = pd.DataFrame(random_returns).transpose().cumsum() * 100
        cur_ret = data["returns"].loc[str(start_year):str(end_year)].cumsum() * 100
    else:
        # Compound returns for alternative method
        df_ret = ((1 + pd.DataFrame(random_returns).transpose()).cumprod() - 1) * 100
        cur_ret = ((1 + data["returns"].loc[str(start_year):str(end_year)]).cumprod() - 1) * 100
    
    # Ensure that the time index aligns with the data time index
    date_range = data.loc[str(start_year):str(end_year)].index  # Correct time index

    # Resample the percentile arrays to match the length of df_ret.index
    p_90 = pd.Series(np.percentile(df_ret, 90, axis=1), index=date_range)
    p_50 = pd.Series(np.percentile(df_ret, 50, axis=1), index=date_range)
    p_10 = pd.Series(np.percentile(df_ret, 10, axis=1), index=date_range)
    
    # Interpolation if necessary
    f_90 = interp1d(range(len(p_90)), p_90, fill_value="extrapolate")
    f_50 = interp1d(range(len(p_50)), p_50, fill_value="extrapolate")
    f_10 = interp1d(range(len(p_10)), p_10, fill_value="extrapolate")
    p_90 = f_90(np.linspace(0, len(p_90) - 1, len(df_ret.index)))
    p_50 = f_50(np.linspace(0, len(p_50) - 1, len(df_ret.index)))
    p_10 = f_10(np.linspace(0, len(p_10) - 1, len(df_ret.index)))
    
    # Plotting the results
    plt.figure(figsize=(20, 8))
    plt.plot(date_range, p_90, color="#39B3C7", label="90th Percentile")
    plt.plot(date_range, p_50, color="#39B3C7", label="50th Percentile")
    plt.plot(date_range, p_10, color="#39B3C7", label="10th Percentile")
    plt.plot(cur_ret.index, cur_ret, color="blue", alpha=0.60, linewidth=3, label="Current returns")
    plt.fill_between(date_range, p_90, p_10, where=(p_90 > p_10), color="#669FEE", alpha=0.20, label="Monte Carlo area")
    
    plt.ylabel("Cumulative returns %", size=13)
    plt.title("MONTE CARLO SIMULATION", size=20)
    plt.legend()
    plt.show()
```

```python
# Run the Monte Carlo simulation
monte_carlo(run_tp_sl_output, method="simple", start_year=2012, end_year=2021, num_simulations=100)
```
